Programa para calcular el valor de la función `exp()` según la serie
de McLaurin
$$ e^x = \sum_{n=0}^\infty \frac{x^n}{n!}$$
El programa debe:
- pedir el valor de x (Opcional: comprobar que $x<1$)
- pedir el valor mínimo del último término a añadir en la suma
- acumular los valores de la suma (podéis usar función `math.factorial()`
- acumular **el número de valores** incluidos en la suma.
- escribir por pantalla el valor aproximado y exacto de la función exp(), además del número de términos de la suma

```python
import math
x = float(input("Introduce el valor de x:"))
min_val = float(input("Introduce el valor mínimo de los términos a añadir"))
s = 0.0
n = 0
term = x**n/math.factorial(n) # siempre es 1
while term > min_val:
    s = s + term
    n = n + 1
    term = x**n/math.factorial(n)
print(f"Valor de la funcion: {math.exp(x):.4f} Serie: {s:.4f} / {n-1} Iteraciones") 
```
