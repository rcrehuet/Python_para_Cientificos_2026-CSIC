```python
minimize(f, x0=[5,4])
```

```python
def f(x,y):
    """
    input:
    x : scalar
    y : scalar
    returns a scalar
    """
    return x**2+2*y**2-3
```

```python
def g(x):
    """
    input:
    x : np.array
    returns a numpy float
    """
    return x[0]**2+2*x[1]**2-3
```

```python
f(2,3)
```

```python
g(np.array([2,3]))
```

```python
resultado = minimize(g, x0=[5,4])
```

```python
type(resultado)
```

```python
resultado
```

```python
if resultado["success"]:
    print("optimization sucessful")
    f_opt = resultado["fun"]
    x_opt = resultado.x
else:
    print("optimization not converged")
```
