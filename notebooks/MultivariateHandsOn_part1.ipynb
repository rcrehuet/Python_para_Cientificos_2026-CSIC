#### Between-group Variance and Within-group Variance for a Variable

- [Within-group variance](https://www.statisticshowto.com/within-group-variation/#:~:text=Within%2DGroup%20Variation:%20Definition%20and,SSW:%20Sum%20of%20Squares%20Within.) measures the spread or variability of individual data points inside each group, representing random error or natural differences not explained by the group differences. It represents "noise" or random fluctuations not caused by the independent variables being tested.

- [Between-group variance](https://www.statisticshowto.com/between-group-variation) measures how much the means of different groups vary from the overall mean, indicating differences attributable to the grouping factor (like a treatment or condition). It's the "*explained*" variation, showing if group averages are truly different from each other, contrasting with within-group variance (random error). 

A large between-group variance relative to within-group variance suggests the factor distinguishing the groups has a significant effect. 

**Calculate within-group variance:**

$$ \frac{\sum_{i=1}^g \sum_{j=1}^{g_{samp}} \left ( X_{ij} - {\overline{X_i}} \right )^2}{N - g}$$
 
where $g$ is the number of groups, $N$ the total number of data points,  $g_{samp}$ is the number of data points belonging to each group, $X_{ij}$ is the value of an individual data point belonging to group $i$, ${\overline{X_i}}$ is the mean value of X in group $i$

If we want to calculate the within-groups variance for a particular variable (for example, ''V2''), we can use the function below:

```python
groups = sorted(set(y)) # a python set withe unique values in y (i.e. V1 column)
ngroups = len(groups);
```

```python
#Compute variance within the group
numtotal = 0
denomtotal = 0
for g in groups:
    g_data = X.V2[y == g]
    l = len(g_data)
    #variance of the elements belonging to the same group
    var_g = np.var(g_data) #g_data.var() Population variance vs Sample Variance
    numg = l*var_g #weighted by the number of samples
    denomg= l
    numtotal = numtotal + numg
    denomtotal = denomtotal + denomg
Vw = numtotal / (denomtotal - ngroups)
Vw
```

**Calculate between-group variance:**

$$ \frac{\sum_{i=1}^g g_{samp} \left (  {\overline{X}} - {\overline{X_i}} \right )^2}{g - 1}$$

If we want to calculate the within-groups variance for a particular variable (for example, ''V2''), we can use the function below:

```python
ss = []
mean_overall = np.mean(X.V2) #overall mean
for g in groups:
    g_data = X.V2[y == g]
    l = len(g_data)
    #mean value in the group
    mean_g = np.mean(g_data)
    #variation to the overall mean
    numg = (mean_overall - mean_g)**2 
    ss.append(l*numg) #weighted by the number of samples
# calculate the between-groups variance
Vb = sum(ss) / (ngroups - 1)
Vb
```

We can calculate the **separation achieved by the V2 variable** as its between-groups variance divided by its within-groups variance. Thus, the separation achieved by V2 is calculated as:

```python
Vb/Vw
```

In order to be more sistematic, we can shift the previous calculations of both variances to user-defined functions:

```python
def calc_var_within_group(variable, groups):
    #Compute variance within the group
    numtotal = 0
    denomtotal = 0
    for g in groups:
        g_data = variable[y == g]
        l = len(g_data)
        #variance of the elements belonging to the same group
        var_g = np.var(g_data) #g_data.var() Population variance vs Sample Variance
        numg = l*var_g #weighted by the number of samples
        denomg= l
        numtotal = numtotal + numg
        denomtotal = denomtotal + denomg
    Vw = numtotal / (denomtotal - ngroups)
    return Vw
```

```python
def calc_var_between_group(variable, groups):
    #Compute variance between the groups
    ss = []
    mean_overall = np.mean(variable) #overall mean
    for g in groups:
        g_data = variable[y == g]
        l = len(g_data)
        #mean value in the group
        mean_g = np.mean(g_data)
        #variation to the overall mean
        numg = (mean_overall - mean_g)**2 
        ss.append(l*numg) #weighted by the number of samples
    # calculate the between-groups variance
    Vb = sum(ss) / (ngroups - 1)
    return Vb
```

```python
calc_var_within_group(X.V2,groups)
```

```python
calc_var_between_group(X.V2, groups)
```

The *separation* introduced by V2 in the group is calculated as the ratio of both variances:

$$ sep = \frac{Vb}{Vw} $$

```python
calc_var_between_group(X.V2, groups)/calc_var_within_group(X.V2,groups)
```

If we want now to calculate the separations achieved by all of the variables in a multivariate data set, we can run the both functions for each independent variable:

```python
for V in X: #for each column in the independent variables data set
        Vw = calc_var_within_group(X[V],groups)
        Vb = calc_var_between_group(X[V], groups)
        sep = Vb/Vw
        print(f"var: {V:3s} Vw= {Vw:10.4f} \t Vb={Vb:10.4f} \t separation={sep:10.4f}")
```

Let's have a look at the table. **The individual ''V'' variable which gives the greatest separations between the groups (the wine varieties) is ''V8'' (separation 230.7).** However, keep in mind that this is a *multidimensional data set*. While V8 is the best single variable to discriminate data, there may be a linear combination of the individual variables that might give a larger separation between the groups.

#### Between-groups Covariance and Within-groups Covariance for Two Variables

See notes on covariance:

$$\sigma_{xy} = \sum_{j=1}^N (x_j - {\overline{x}})(y_j - {\overline{y}})$$

In our case, we have within-group covariance

$$\sigma_{xy} = \frac{\sum_{i=1}^g \sum_{j=1}^{g_{samp}} (x_{ij} - {\overline{x_i}})(y_{ij} - {\overline{y_i}})}{N - g}$$

```python
def Covariance_within_Group(var1, var2, groups):
    Covw = 0.0
    for g in groups:
        cc = []
        g_data1 =  var1[ y == g ]
        g_data2 =  var2[ y == g ]
        g_mean1 = np.mean(g_data1)
        g_mean2 = np.mean(g_data2)
        cc = [(d1 - g_mean1)*(d2 - g_mean2) for d1,d2 in zip(g_data1, g_data2)]
        cov_group = sum(cc)
        Covw = Covw + cov_group
    Covw = Covw / (len(var1) - ngroups)
    return Covw                        
```

Covariance within group for variables V8 and V11:

```python
Covariance_within_Group(X.V8, X.V11, groups)
```

In the case of the between-groups covariance:


$$ \frac{\sum_{i=1}^g g_{samp} \left (  {\overline{X}} - {\overline{X_i}} \right ) \left (  {\overline{Y}} - {\overline{Y_i}} \right )}{g - 1}$$

```python
def  Covariance_between_Groups(var1, var2, groups):
    Covb = []
    #calculate overall means
    mean1 = np.mean(var1)
    mean2 = np.mean(var2)
    #between groups covariance
    for g in groups:
        g_data1 =  var1[ y == g ]
        g_data2 =  var2[ y == g ]
        g_mean1 = np.mean(g_data1)
        g_mean2 = np.mean(g_data2)
        Covb.append((g_mean1 - mean1)*(g_mean2 - mean2)*len(g_data1))
    Covb = sum(Covb) / (ngroups - 1)
    return Covb                        
```

```python
Covariance_between_Groups(X.V8, X.V11, groups)
```

Thus, for V8 and V11, the between-groups covariance is -60.71 and the within-groups covariance is 0.29. Since the within-groups covariance is positive (0.29), it means V8 and V11 are positively related within groups: for individual data points from the same group, individuals with a high value of V8 tend to have a high value of V11, and vice versa. Since the between-groups covariance is negative (-60.71), V8 and V11 are negatively related between groups: groups with a high mean value of V8 tend to have a low mean value of V11, and vice versa.

It is often interesting to investigate whether any of the variables in a multivariate data set are significantly correlated. To calculate the linear (Pearson) correlation coefficient ([PCC](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient)) between all the variables in the data set we can used the `DataFrame.corr()` method from Pandas and get the **correlation matrix**:

The matrix can be rapidly inspected and we can identify larger elements such as V8/V7 (0.79) or V11/V12 (-0.52, anticorrelation). A nice way to do it is graphically using the tools provided by Seaborn:

```python
sns.heatmap(corr_matrix, vmax=1., square=False).xaxis.tick_top()
```

A very elegant alternative visualization is via a [Hinton diagram](https://matplotlib.org/stable/gallery/specialty_plots/hinton_demo.html). The color of the boxes determines the sign of the correlation, in this case red for positive and blue for negative correlations; while the size of the boxes determines their magnitude, the bigger the box the higher the magnitude.

```python
# adapted from http://matplotlib.org/examples/specialty_plots/hinton_demo.html
# adapted from https://python-for-multivariate-analysis.readthedocs.io/a_little_book_of_python_for_multivariate_analysis.html#between-groups-covariance-and-within-groups-covariance-for-two-variables
def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a correlation matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2**np.ceil(np.log(np.abs(matrix).max())/np.log(2))

    ax.patch.set_facecolor('lightgray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'blue' if w > 0 else 'red'
        size = np.sqrt(np.abs(w))
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    nticks = matrix.shape[0]
    ax.xaxis.tick_top()
    ax.set_xticks(range(nticks))
    ax.set_xticklabels(list(matrix.columns), rotation=90)
    ax.set_yticks(range(nticks))
    ax.set_yticklabels(matrix.columns)
    ax.grid(False)

    ax.autoscale_view()
    ax.invert_yaxis()

hinton(corr_matrix)
```

Large blue squares identify variables that correlate strongly and positively while large red squares are found the air of variables have a strong negative correlation.
