#### Between-group Variance and Within-group Variance for a Variable

- [Within-group variance](https://www.statisticshowto.com/within-group-variation/#:~:text=Within%2DGroup%20Variation:%20Definition%20and,SSW:%20Sum%20of%20Squares%20Within.) measures the spread or variability of individual data points inside each group, representing random error or natural differences not explained by the group differences. It represents "noise" or random fluctuations not caused by the independent variables being tested.

- [Between-group variance](https://www.statisticshowto.com/between-group-variation) measures how much the means of different groups vary from the overall mean, indicating differences attributable to the grouping factor (like a treatment or condition). It's the "*explained*" variation, showing if group averages are truly different from each other, contrasting with within-group variance (random error). 

A large between-group variance relative to within-group variance suggests the factor distinguishing the groups has a significant effect. 

**Calculate within-group variance:**

$$ \frac{\sum_{i=1}^g \sum_{j=1}^{g_{samp}} \left ( X_{ij} - {\overline{X_i}} \right )^2}{N - g}$$
 
where $g$ is the number of groups, $N$ the total number of data points,  $g_{samp}$ is the number of data points belonging to each group, $X_{ij}$ is the value of an individual data point belonging to group $i$, ${\overline{X_i}}$ is the mean value of X in group $i$

If we want to calculate the within-groups variance for a particular variable (for example, ''V2''), we can use the function below:

```python
groups = sorted(set(y)) # a python set withe unique values in y (i.e. V1 column)
ngroups = len(groups);
```

```python
#Compute variance within the group
numtotal = 0
denomtotal = 0
for g in groups:
    g_data = X.V2[y == g]
    l = len(g_data)
    #variance of the elements belonging to the same group
    var_g = np.var(g_data) #g_data.var() Population variance vs Sample Variance
    numg = l*var_g #weighted by the number of samples
    denomg= l
    numtotal = numtotal + numg
    denomtotal = denomtotal + denomg
Vw = numtotal / (denomtotal - ngroups)
Vw
```

**Calculate between-group variance:**

$$ \frac{\sum_{i=1}^g g_{samp} \left (  {\overline{X}} - {\overline{X_i}} \right )^2}{g - 1}$$

If we want to calculate the within-groups variance for a particular variable (for example, ''V2''), we can use the function below:

```python
ss = []
mean_overall = np.mean(X.V2) #overall mean
for g in groups:
    g_data = X.V2[y == g]
    l = len(g_data)
    #mean value in the group
    mean_g = np.mean(g_data)
    #variation to the overall mean
    numg = (mean_overall - mean_g)**2 
    ss.append(l*numg) #weighted by the number of samples
# calculate the between-groups variance
Vb = sum(ss) / (ngroups - 1)
Vb
```

We can calculate the **separation achieved by the V2 variable** as its between-groups variance divided by its within-groups variance. Thus, the separation achieved by V2 is calculated as:

```python
Vb/Vw
```

In order to be more sistematic, we can shift the previous calculations of both variances to user-defined functions:

```python
def calc_var_within_group(variable, groups):
    #Compute variance within the group
    numtotal = 0
    denomtotal = 0
    for g in groups:
        g_data = variable[y == g]
        l = len(g_data)
        #variance of the elements belonging to the same group
        var_g = np.var(g_data) #g_data.var() Population variance vs Sample Variance
        numg = l*var_g #weighted by the number of samples
        denomg= l
        numtotal = numtotal + numg
        denomtotal = denomtotal + denomg
    Vw = numtotal / (denomtotal - ngroups)
    return Vw
```

```python
def calc_var_between_group(variable, groups):
    #Compute variance between the groups
    ss = []
    mean_overall = np.mean(variable) #overall mean
    for g in groups:
        g_data = variable[y == g]
        l = len(g_data)
        #mean value in the group
        mean_g = np.mean(g_data)
        #variation to the overall mean
        numg = (mean_overall - mean_g)**2 
        ss.append(l*numg) #weighted by the number of samples
    # calculate the between-groups variance
    Vb = sum(ss) / (ngroups - 1)
    return Vb
```

```python
calc_var_within_group(X.V2,groups)
```

```python
calc_var_between_group(X.V2, groups)
```

The *separation* introduced by V2 in the group is calculated as the ratio of both variances:

$$ sep = \frac{Vb}{Vw} $$

```python
calc_var_between_group(X.V2, groups)/calc_var_within_group(X.V2,groups)
```

If we want now to calculate the separations achieved by all of the variables in a multivariate data set, we can run the both functions for each independent variable:

```python
for V in X: #for each column in the independent variables data set
        Vw = calc_var_within_group(X[V],groups)
        Vb = calc_var_between_group(X[V], groups)
        sep = Vb/Vw
        print(f"var: {V:3s} Vw= {Vw:10.4f} \t Vb={Vb:10.4f} \t separation={sep:10.4f}")
```

Let's have a look at the table. **The individual ''V'' variable which gives the greatest separations between the groups (the wine varieties) is ''V8'' (separation 230.7).** However, keep in mind that this is a *multidimensional data set*. While V8 is the best single variable to discriminate data, there may be a linear combination of the individual variables that might give a larger separation between the groups.

#### Between-groups Covariance and Within-groups Covariance for Two Variables

See notes on covariance:

$$\sigma_{xy} = \sum_{j=1}^N (x_j - {\overline{x}})(y_j - {\overline{y}})$$

In our case, we have within-group covariance

$$\sigma_{xy} = \frac{\sum_{i=1}^g \sum_{j=1}^{g_{samp}} (x_{ij} - {\overline{x_i}})(y_{ij} - {\overline{y_i}})}{N - g}$$

```python
def Covariance_within_Group(var1, var2, groups):
    Covw = 0.0
    for g in groups:
        cc = []
        g_data1 =  var1[ y == g ]
        g_data2 =  var2[ y == g ]
        g_mean1 = np.mean(g_data1)
        g_mean2 = np.mean(g_data2)
        cc = [(d1 - g_mean1)*(d2 - g_mean2) for d1,d2 in zip(g_data1, g_data2)]
        cov_group = sum(cc)
        Covw = Covw + cov_group
    Covw = Covw / (len(var1) - ngroups)
    return Covw                        
```

Covariance within group for variables V8 and V11:

```python
Covariance_within_Group(X.V8, X.V11, groups)
```

In the case of the between-groups covariance:


$$ \frac{\sum_{i=1}^g g_{samp} \left (  {\overline{X}} - {\overline{X_i}} \right ) \left (  {\overline{Y}} - {\overline{Y_i}} \right )}{g - 1}$$

```python
def  Covariance_between_Groups(var1, var2, groups):
    Covb = []
    #calculate overall means
    mean1 = np.mean(var1)
    mean2 = np.mean(var2)
    #between groups covariance
    for g in groups:
        g_data1 =  var1[ y == g ]
        g_data2 =  var2[ y == g ]
        g_mean1 = np.mean(g_data1)
        g_mean2 = np.mean(g_data2)
        Covb.append((g_mean1 - mean1)*(g_mean2 - mean2)*len(g_data1))
    Covb = sum(Covb) / (ngroups - 1)
    return Covb                        
```

```python
Covariance_between_Groups(X.V8, X.V11, groups)
```

Thus, for V8 and V11, the between-groups covariance is -60.71 and the within-groups covariance is 0.29. Since the within-groups covariance is positive (0.29), it means V8 and V11 are positively related within groups: for individual data points from the same group, individuals with a high value of V8 tend to have a high value of V11, and vice versa. Since the between-groups covariance is negative (-60.71), V8 and V11 are negatively related between groups: groups with a high mean value of V8 tend to have a low mean value of V11, and vice versa.

A very elegant alternative visualization is via a [Hinton diagram](https://matplotlib.org/stable/gallery/specialty_plots/hinton_demo.html). The color of the boxes determines the sign of the correlation, in this case red for positive and blue for negative correlations; while the size of the boxes determines their magnitude, the bigger the box the higher the magnitude.

```python
# adapted from http://matplotlib.org/examples/specialty_plots/hinton_demo.html
# adapted from https://python-for-multivariate-analysis.readthedocs.io/a_little_book_of_python_for_multivariate_analysis.html#between-groups-covariance-and-within-groups-covariance-for-two-variables
def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a correlation matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2**np.ceil(np.log(np.abs(matrix).max())/np.log(2))

    ax.patch.set_facecolor('lightgray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'blue' if w > 0 else 'red'
        size = np.sqrt(np.abs(w))
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    nticks = matrix.shape[0]
    ax.xaxis.tick_top()
    ax.set_xticks(range(nticks))
    ax.set_xticklabels(list(matrix.columns), rotation=90)
    ax.set_yticks(range(nticks))
    ax.set_yticklabels(matrix.columns)
    ax.grid(False)

    ax.autoscale_view()
    ax.invert_yaxis()

hinton(corr_matrix)
```

Large blue squares identify variables that correlate strongly and positively while large red squares are found the air of variables have a strong negative correlation.

### Standardising Variables

If we want to compare different variables that have different units, and therefore very different variances, it is a good idea to **first standardise the variables**.
In our example, we found above that the concentrations of the 13 chemicals in the wine samples show a wide range of standard deviations, from 0.124 for V9 (variance 0.015) to 314.022 for V14 (variance 98609.60). In this case, it is not a good idea to use the unstandardised chemical concentrations as the input for a principal component analysis (PCA, see below) of the wine samples, as if we did that, the first principal component would be dominated by the variables which show the largest variances, such as V14.

Therefore, it is highhly advisable to first standardise the variables so that **they all have variance 1 and mean 0**, and to then carry out the principal component analysis on the standardised data. This would allow us to find the principal components that provide the best low-dimensional representation of the variation in the original data, without being overly biased by those variables that show the most variance in the original data.



#### Standardization using Python

Standardization, also known as Z-score normalization, involves transforming the data such that it has a mean of zero and a standard deviation of one. Mathematically, the standardization formula for a feature 𝑥 is given by:

$$ z = \frac{x - \mu}{\sigma}$$

where
- 𝑧 is the standardized value,
- 𝑥 is the original value,
- $\mu$ is the mean and,
- $\sigma$ is the standard deviation.

## Principal Component Analysis

PCA (Principal Component Analysis) is a dimensionality reduction technique and helps us to reduce the number of features in a dataset while keeping the most important information. It changes complex datasets by transforming correlated features into a smaller set of uncorrelated components. 

PCA uses linear algebra to transform data into new features called principal components. It finds these by calculating eigenvectors (directions) and eigenvalues (importance) from the covariance matrix. PCA selects the top components with the highest eigenvalues and projects the data onto them simplify the dataset. It prioritizes the directions where the data varies the most because more variation means more useful information.

In our example, we have 13 chemical concentrations describing wine samples from three different varieties. We can carry out PCA to see if we can get most of the variation between samples using a smaller set of variables (these will be our *principal components*). These new variables will be a linear combination of all or some of the 13 chemical concentrations.

Typically, the first step is often to **standardise the variables** under study function. We just did that in the previous section (see above). 

Once we have standardised our variables, we can carry out a principal component analysis using the **[`PCA` class](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA)** from `sklearn.decomposition`  package and its [`fit` method](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA.fit). This method will fit the model with our standardised data stored in variable `standardised_X`. The default solver is [**Singular Value Decomposition**](https://en.wikipedia.org/wiki/Singular_value_decomposition) (“svd”). Keep in mind that `PCA().fit()` calculates the necessary parameters (eigenvectors and eigenvalues) to define the new space storing them within the fitted PCA object, but **it does not return the transformed dataset**. For more information you can type help(PCA) in the python console.

#### Deciding how many Principal Components to retain
So far we have reduced nothing, we had 13 variables (V2 to V14) in the original data, we have built 13 standardised variable (Z2 to Z14) and the PCA procedure has yielded 13 principal components (PC1 to PC13). However, we just saw above that we have now a hierarchy and some components explain most of the variance in our data. This allows us to reduce the dimensionality of our analysis. In order to decide how many principal components should be retained, it is common to summarise the results of a principal components analysis by making a [scree plot](https://en.wikipedia.org/wiki/Scree_plot). This consists of a simple line plot of the variance values:

```python
yy = pca.explained_variance_
x = np.arange(1,len(pca.explained_variance_)+1)
plt.plot(x, yy, "o-")
plt.xticks(x, pc_labels, rotation=60)
plt.ylabel("Variance")
plt.grid()
plt.show()
```

According to the literature several criteria can be applied to selected the reduced set of PCs to proceed with the analysis:
- One possible criterion is to observe that a change in slope in the scree plot occurs at component 4. Therefore, it could be argued that the first three components should be retained.
- Another way of deciding how many components to retain is to use *Kaiser’s criterion*: only retain principal components for which the variance is above 1 (standardised data!!). We would retain, again, the first three principal components.
- A third way to decide how many PCs to retain is to decide to keep the number of components required to explain at least some minimum amount of the total variance. For example, to explain at least 80% of the variance, we would retain the first five principal components, as we can see from the table we computed above.

## Linear Discriminant Analysis
The purpose of principal component analysis is to find the best low-dimensional representation of the variation in a multivariate data set. In our wine data set example, we have 13 chemical concentrations describing wine samples from three wine varieties. By carrying out a PCA, we have found that most of the variation in the chemical concentrations between the samples can be captured using the first two principal components, PC1 and PC2. Each of the principal components is a particular linear combination of the 13 chemical concentrations.

The purpose of linear discriminant analysis (LDA) is to find the linear combinations of the original variables (the 13 chemical concentrations here) that gives the best possible separation between the groups in the data set. Notice the difference with PCA where we also seek linear combinations. In LDA we aim for the best group separation, we rely on having previously labelled the data points in groups. In PCA the linear combination seeks the largest contributions to the variance without any knowledge of the labelling.

The wines in our data set come from three different varieties, so the number of groups (G) is 3, and the number of variables is 13 (13 chemicals’ concentrations; p = 13). The maximum number of useful discriminant functions that can separate the wines by cultivar is the minimum of G-1 and p, and so in this case it is the minimum of 2 and 13, which is 2. Thus, we can find at most 2 useful discriminant functions to separate the wines by cultivar, using the 13 chemical concentration variables.

### Decission rules and Missclassification rate

In LDA, decision rules (also called allocation rules or classification rules) specify how a **new observation** is assigned to one of the predefined classes after the discriminant functions have been constructed. In our example, it would mean how would we assign a new wine sample (with 13 chemical species concentrations) to a given wine variety: 1, 2, or 3. In the context of LDA, this procedure has two main steps:

- Learning step. We have already completed this step. From labeled data, LDA finds one or more discriminant functions that best separate the classes. We have LD1 and LD2.

- Decision step. Given a new observation, the allocation rule decides which class it belongs to.


The basic idea is that once once data are projected onto the LDA space, each class occupies a region. The allocation rule assigns a new point to the class for which it is “most compatible”. This compatibility can be defined in several ways:
- maximum discriminant score
- minimum distance to class mean
- maximum posterior probability



We can calculate the mean values of the discriminant functions for each of the three wine varieties:

and standard deviation:

Looking at the numbers we find that the mean value of LD1 is -3.422 for variety 1, -0.0797 for variety 2, and 4.3245 for group 3. The mid-way point between the mean values for groups 1 and 2 is -1.751, and the mid-way point between the mean values for groups 2 and 3 is 2.122.

We can therefore use the following **decision rule**:
- if $LD_1 \le -1.751$ $\Rightarrow$ predict the wine sample to be from variety 1
- if $2.122 \ge LD_1 \gt -1.751$ $\Rightarrow$  predict the sample to be from variety 2
- if $LD_1 \gt 2.122$ $\Rightarrow$  predict the sample to be from variety 3

We can check the reliability of this criterion using our wine data set:

```python
conf_m = np.zeros((3,3))
cutoff=[-1.751, 2.122]
for i,g in enumerate(groups):
    g_data = lda_transformed_data[lda_transformed_data.Group==g]
    g1 = (g_data.LD1 < cutoff[0]).sum()
    conf_m[i,0] = g1
    print(f"Number of samples from group {g} predicted as group 1 is: {g1}")
    g2 = ((cutoff[1] > g_data.LD1) & (g_data.LD1 >= cutoff[0])).sum()
    print(f"Number of samples from group {g} predicted as group 2 is: {g2}")
    conf_m[i,1] = g2
    g3 = (g_data.LD1 >= cutoff[1]).sum()
    print(f"Number of samples from group {g} predicted as group 3 is: {g3}")
    conf_m[i,2] = g3
    print("===")
```

This information can be displayed in what is called a [*confusion matrix*](https://en.wikipedia.org/wiki/Confusion_matrix):

```python
print("Confusion Matrix:")
print(conf_m)
```

There are 3+6+1=10 wine samples that are misclassified, out of 177 wine samples. According to our LDA and prediction criteria 3 samples from variety 1 are assigned to group 2, 6 samples from variety 2 are to group 1, and 1 is predicted to be from group 3. All the samples from variety 3 are classified correctly. Therefore, the misclassification rate is 5.6% (10 out of 177). This a rather low value, and therefore the allocation rule appears to be relatively realible. However, this is probably an underestimate of the misclassification rate, given that we have used the *training set* as the *testing set*. If we calculated the misclassification rate for a separate test set consisting of data other than that used to make the allocation rule, we would probably get a higher estimate of the misclassification rate.
